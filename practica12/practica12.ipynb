### Actividad 1 

Implementar el algoritmo de Karatsuba que usa la técnica divide y vencerás para realizar la multiplicación de enteros grandes.

```python
#Francisco Abimael Oro Estrada
def mul(u = 1,v = 1):
    utemp = len(str(u))
    vtemp = len(str(v))
    n = utemp if utemp > vtemp else vtemp
    if n==1:
        return v*u
    else: 
        s=int(n/2)
        w = u // 10 ** s
        x = u % 10 ** s
        y = v // 10 ** s
        z = v % 10 ** s
        return (10 ** (2*s) * mul(w, y) + (10**s)*mul(w,z) + mul(x,y) + mul(x, z))
    
if __name__ == '__main__':
    print(mul(21389,212379))
```

- 2.1 Implementar en Python 3 el algoritmo de Merge sort que utiliza la técnica Divide y Vencerás. Además graficar una función tabular T(n) (n vs t) como la mostrada abajo. Es decir ordenar secuencias de diferente tamaño y obtener el tiempo que se tarda en ordenarlas y graficar esos datos. 

```python
def merge(A, p, q, r):
    izq = A[p:q+1]
    der=A[q+1:r+1]
    i=0
    j=0
    for k in range(p, r+1):
        if j>= len(der) or ( i < len(izq) and izq[i] < der[j]):
            A[k] = izq[i]
            i += 1
        else:
            A[k] = der[j]
            j += 1

def mergeSort(A, p, r):
    if p < r:
        q = int((p+r)/2)
        mergeSort(A, p, q)
        mergeSort(A, q+1, r)
        merge(A, p, q, r)

if __name__ == '__main__':
    arr = [int(input("Ingrese un numero : ")) for i in range(6)]
    mergeSort(arr, 0, len(arr)-1)
    print(arr)
```

- 2.2 Obtener una gráfica similar pero con el algoritmo de ordenamiento seleccion sort visto en la práctica 10 y comparar ambas gráficas.

```python
import random
from time import time
import matplotlib.pyplot as plt

def selection_sort(array):  
    length = len(array)  
      
    for i in range(length-1):  
        minIndex = i  
          
        for j in range(i+1, length):  
            if array[j]<array[minIndex]:  
                minIndex = j  
                  
        array[i], array[minIndex] = array[minIndex], array[i]  
          
          
    return array      

def graficas():
    tm = []
    tm2 = []
    nD = [k*100 for k in range(1, 31)]
    for k in nD:
        datosM = random.sample(range(1, 1000000000), k)
        
        t0 = time()
        mergeSort(datosM, 0, len(datosM)-1)
        t1 = time()
        tm.append(round(t1-t0, 6))

    for k in nD:
        datosM = random.sample(range(1, 1000000000), k)
        
        t0 = time()
        selection_sort(datosM)
        t1 = time()
        tm2.append(round(t1-t0, 6))
        
    fig, ax = plt.subplots()
    ax.plot(nD, tm, label = "Merge Sort", marker = "*", color="r")
    ax.plot(nD, tm2, label = "Merge Sort", marker = "*", color="b")
    
    ax.set_xlabel("n Datos")
    ax.set_ylabel("Tiempo")
    ax.grid(True)
        
    plt.show()
```

### Actividad 3 
Resolver el problema de calculo de los numeros de la sucesion de fibonacci con programacion dinamica Top-down y bottom-up 

```python
# Dynamic programming implementation 
class fibo:

    def top_down(self, n: int)->int:
        # An array is needed as memorization will be used in the top-down implementation.
        # It's mandatory to initialize this array with a value that doesn't actually exist in the fibo sequence
        self.memo = [-1 for i in range(n+1)]
        return self.top_down_aux(n)
    
    # Case I. Top-down implementation:
    def top_down_aux(self, n):
        # Base cases are fib(1) and fib(0)
        if n == 0 or n==1:
            return n
        # Return the stored value in case the problem has been previously solved 
        if self.memo[n] != -1:
            return self.memo[n]
        # Return every element in the sequence and store it in the memory array
        self.memo[n] = self.top_down_aux(n-1) + self.top_down_aux(n-2)
        return self.memo[n]
    
    # Case II. Bottom-up implementation
    def bottom_up(self, n:int)->int:
        # Base cases are fib(0) and fib(1)
        base = [0, 1]
        # Note that the loop starts from 2 as the base cases are already stored in the base array
        for i in range(2, n+1):
            # Each new term in the sequence is inserted into the base array
            base.append(base[i-1] + base[i-2])
        # Return the greatest term in the array, as required by the user
        return base[n]

if __name__ == '__main__':
    fib = fibo()
    print(f"Using the top_down implementation : {fib.top_down(11)}")
    print(f"Using the bottom_up implementation : {fib.bottom_up(11)}")
```

### Actividad 4 
Resolver el problema del cambio con programación dinámica bottom-up.


```python
def coinChange(coins: list[int], count: int) -> int:
    # Use an array of count + 1 items
    # Assign the maximum possible value for each element in the array
    dp = [count + 1 for i in range(count + 1)]
    # Set the base case when the count is empty, so there won't be any combination of coins to reach it
    dp[0] = 0


    for i in range(1, count + 1):
        for coin in coins:
            # The difference must be greater or equal than 0, as negative counts will be neglected
            if i - coin >= 0:
                """
                Notice :
                (1) That 1 comes from the known values of dp for the coins existing in the list provided as 'coins', namely:
                    for the set of coins = [1, 2, 5, 10], we know that it takes exactly one coin to reach any element from that set
                (2) dp[i - coin] stands for the minimum number of combinations for the residual
                """
                dp[i] = min(dp[i], 1 + dp[i - coin])
    return dp[count] if dp[count] != count + 1 else  -1




if __name__ == '__main__':
    coins = [1, 2, 5, 10]
    number = int(input("Enter a number : "))
    print(f"The minimum number of combinations to reach that quantity is {coinChange(coins, number)}")
```
